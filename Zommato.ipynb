```python
#Figure out how much amount has been spend on the restuarants for various countires.
final_zdata['Average Cost for two'].groupby(final_zdata['Country']).mean().sort_values(ascending=False).plot(kind='bar')
```

Observation: For above plot, we can see that the average cost of food in 'Indonesia' is high.
    
     In the given data, each country's average cost is mentioned in their respective currency. Lets standardize the currency unit to Dollar. Below conversion rates are according to Google on the day of writing this code

```python
conversion_rates= {'Botswana Pula(P)':0.018, 'Brazilian Real(R$)':0.19,'Dollar($)':1,'Emirati Diram(AED)':0.27,
    'Indian Rupees(Rs.)':0.012117,'Indonesian Rupiah(IDR)':0.000066,'NewZealand($)':0.632550,'Pounds(ÂŒÂ£)':1.21,
    'Qatari Rial(QR)':0.27,'Rand(R)':0.056,'Sri Lankan Rupee(LKR)':0.0027,'Turkish Lira(TL)':0.053}
```

```python
final_zdata['New cost']=final_zdata['Average Cost for two']*final_zdata['Currency'].map(conversion_rates)
```

```python
final_zdata['New cost'].groupby(final_zdata['Country']).mean().sort_values(ascending=False).plot(kind='bar')
```

We can see that after bringing all country's cost into Dollar, we can observe that Singapore has the expensive food while Turkey has the least expensive food.

# Feature Engineering

```python
sns.boxplot(numericcols)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
numeric_robust=RobustScaler().fit_transform(numericcols.drop('Aggregate rating',axis=1))
```

```python
numeric_robust=pd.DataFrame(numeric_robust,columns=['Price range','New cost','Votes'])
```

```python
numeric_robust.head()
```

```python
object_dummy=pd.get_dummies(objectcols,columns=objectcols.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
objectcols_labelencode=objectcols.apply(le.fit_transform)
```

```python
objectcols_labelencode.head()
```

```python
object_dummy.head()
```

```python
print(objectcols.shape)
print(object_dummy.shape)
print(objectcols_labelencode.shape)
```

```python
combined_df=pd.concat([numeric_robust,objectcols_labelencode],axis=1)
```

```python
combined_df['Aggregate rating']=numericcols['Aggregate rating']
```

```python
#Splitting into X and y
X=combined_df.drop(['Aggregate rating'],axis=1)
y=combined_df['Aggregate rating']
```

```python
y.plot(kind='box',vert=False)
```

Observation: We can see there are minimum outliers present in dependent variable 'y'
